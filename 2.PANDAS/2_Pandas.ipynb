# Creating Pandas Series

```python
import pandas as pd
import numpy as np
ser = pd.Series()
data = np.array(['g', 'p', 'r', 'e', 'c'])
ser = pd.Series(data)
print("Pandas Series:\n", ser)
```

```python
import numpy as np
import pandas as pd
info= np.array(['p','a','n','d','a','s'])
print(info)
ser=pd.Series(info)
print(ser)
```

# Accessing data from Series

```python
import pandas as pd
import numpy as np
data = np.array(['g','p','r','e','c','k', 'u','r','n','o','o','l'])
ser = pd.Series(data)
print(ser[:5])
```

# Giving names to index

```python
import pandas as pd
data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}
df = pd.DataFrame(data, index = ["day1", "day2", "day3"])
print(df)
```
